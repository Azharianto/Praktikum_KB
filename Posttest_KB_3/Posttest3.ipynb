```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('bmw.csv')
```

Membuat grafik batang

```python
data_bar = data['model'].value_counts()
plt.figure(figsize=(10,8))
plt.bar(x=data_bar.index, height=data_bar.values, color=["red", "yellow", "blue"])
```

Membuat Histogram

```python
data = pd.read_csv('bmw.csv')
```

```python
plt.hist(data['price'], bins=5, color="SkyBlue", edgecolor='black',)
plt.xlabel('Harga')
plt.ylabel('Frekuensi')
plt.title('Price')
plt.show()
```

```python

plt.figure(figsize=(6, 6))
data['transmission'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, explode=(0.1, 0.1 , 0.1), shadow = True)
plt.title('Distribusi Data berdasarkan transmission')
plt.xlabel('Automatic')
plt.ylabel('Persentase')
plt.show()
```

Membuat Heat map

```python
correlation_matrix = data.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Antar Fitur')
plt.show()
```
